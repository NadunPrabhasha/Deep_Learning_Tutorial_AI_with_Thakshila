```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense  #fully connected layer
```

```python
model=Sequential() #empty NN
```

```python
model.add(Dense(16,input_dim=4,activation='relu')) #adding layer 1(1st hidden layer)
model.add(Dense(40,activation='relu')) #adding layer 2(2nd hidden layer)
model.add(Dense(10,activation='relu')) #adding layer 3(3rd hidden layer)
model.add(Dense(3,activation='softmax')) #adding layer 4(output layer)

model.summary()
```
