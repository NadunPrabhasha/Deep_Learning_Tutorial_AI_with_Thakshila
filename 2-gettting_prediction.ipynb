```python
from keras.models import load_model
```

```python
model = load_model('FFNN-iris.keras')
```

```python
model.summary()
```

```python
import numpy as np

x = np.array([[1.2, 2.4, 4.2, 5.3]], dtype=np.float32)
result = model.predict(x)
print(result)
```

```python
label = np.argmax(result,axis=1)
print(label)
```
