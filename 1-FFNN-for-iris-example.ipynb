```python
import pandas as pd
```

```python
dataset = pd.read_csv(r'D:\Deep_Learning_Tutorial_AI_with_Thakshila\Deep_Learning_Tutorial_codePRO_LK\iris.csv').values
```

```python
print(dataset.shape)
```

```python
print(dataset)
```

```python
data = dataset[:,0:4]
```

```python
print(data)
```

```python
target = dataset[:,4]
```

```python
print(target)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense  #fully connected layer
```

```python
model=Sequential() #empty NN
```

```python
model.add(Dense(16,input_dim=4,activation='relu')) #adding layer 1(1st hidden layer)
model.add(Dense(40,activation='relu')) #adding layer 2(2nd hidden layer)
model.add(Dense(10,activation='relu')) #adding layer 3(3rd hidden layer)
model.add(Dense(3,activation='softmax')) #adding layer 4(output layer)

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
```

```python
from tensorflow.keras.utils import to_categorical
new_target = to_categorical(target)
```

```python
print(new_target)
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.2)
```

```python
print(train_data.shape)
print(test_data.shape)
```

```python
print(test_data)
```

```python
print(test_target)
```

```python
model.fit(
    train_data,
    train_target,
    epochs=1000,
    validation_data=(test_data,test_target),
    batch_size=4) #training
```

```python
from matplotlib import pyplot as plt

plt.plot(model.history.history['loss'],label='loss')
plt.plot(model.history.history['val_loss'],label='val_loss')
plt.legend()
plt.xlabel('# epoch')
plt.ylabel('loss')
```

```python
plt.plot(model.history.history['accuracy'],label='accuracy')
plt.plot(model.history.history['val_accuracy'],label='val_accuracy')
plt.legend()
plt.xlabel('# epoch')
plt.ylabel('accuracy')
```

```python
model.save('FFNN-iris.keras')
```
